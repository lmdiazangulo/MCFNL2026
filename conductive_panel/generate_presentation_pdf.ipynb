```python
# Importar librerías necesarias
from pathlib import Path
import subprocess
```

```python
# Cargar la presentación
tex_path = Path('presentacion_panel_conductivo.tex')
assert tex_path.exists(), f'No existe {tex_path}'
```

```python
# Generar PDF
subprocess.run(['pdflatex', '-interaction=nonstopmode', str(tex_path)], check=True)
```

```python
# Guardar como presentación.pdf
generated_pdf = Path('presentacion_panel_conductivo.pdf')
output_pdf = Path('presentación.pdf')
if generated_pdf.exists():
    output_pdf.unlink(missing_ok=True)
    generated_pdf.replace(output_pdf)
else:
    raise FileNotFoundError(generated_pdf)
print('PDF guardado como', output_pdf.name)
```
